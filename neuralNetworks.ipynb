```python
DATA_PATH = "/content/household_power_consumption.txt"
print("Dataset found:", DATA_PATH)

# neural_networks_simple_mlp.ipynb - Notebook cell content
# Requirements: pandas, numpy, scikit-learn, matplotlib, tensorflow
# Install with: pip install pandas numpy scikit-learn matplotlib tensorflow

import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Input
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# ---------------------------
# USER PARAMETERS (edit if needed)
# ---------------------------
# Assumes this notebook and the txt file are in the same folder
TXT_FILENAME = DATA_PATH # file in same folder as the notebook
INPUT_COL = 'Global_active_power'  # dataset column to use
PAST_STEPS = 30    # number of minutes used as input
FUTURE_STEPS = 120 # minutes to predict
MAX_SAMPLES = None # set to None to use all windows; or an int to limit for speed (e.g., 200000)
TEST_RATIO = 0.15   # chronological test split
VAL_RATIO = 0.15    # chronological validation split
BATCH_SIZE = 128
EPOCHS = 50
RANDOM_SEED = 42
MODEL_SAVE_PATH = 'best_simple_mlp.h5'

# ---------------------------
# 0) reproducibility (best effort)
# ---------------------------
np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

# ---------------------------
# 1) Load the data (file expected in same folder)
# ---------------------------
if not os.path.exists(TXT_FILENAME):
    raise FileNotFoundError(f"Data file not found at {TXT_FILENAME}. Make sure the file is in the same folder as this notebook.")

print("Loading dataset (this may take a while)...")
# The UCI file uses ';' separator, Date and Time columns; missing values marked by '?'
df = pd.read_csv(
    TXT_FILENAME,
    sep=';',
    parse_dates={'datetime': ['Date', 'Time']},
    infer_datetime_format=True,
    low_memory=False,
    na_values=['?', '']
)

# sort by datetime just in case
df.sort_values('datetime', inplace=True)
df.set_index('datetime', inplace=True)

print("Raw shape:", df.shape)
print("Columns:", df.columns.tolist())

# ---------------------------
# 2) Preprocess / cleaning
# ---------------------------
# Convert target column to numeric (it may contain non-numeric tokens)
df[INPUT_COL] = pd.to_numeric(df[INPUT_COL], errors='coerce')

# Resample to ensure 1-minute frequency (UCI dataset already has minute-level timestamps)
df = df.resample('1T').asfreq()

# Check missing ratio for the target
missing_ratio = df[INPUT_COL].isna().mean()
print(f"Missing ratio in {INPUT_COL}: {missing_ratio:.4f}")

# Interpolate short gaps linearly and drop remaining NaNs
# limit=60 -> only interpolate gaps up to 60 minutes; adjust if you prefer
df[INPUT_COL] = df[INPUT_COL].interpolate(method='time', limit=60)
n_before = len(df)
df = df.dropna(subset=[INPUT_COL])
n_after = len(df)
print(f"Dropped {n_before - n_after} rows with missing target after interpolation")

# Work with single feature (Global_active_power). Convert to float32 for efficiency.
series = df[[INPUT_COL]].astype('float32')

# ---------------------------
# 3) Prepare sliding-window indices and splits (chronological)
# ---------------------------
total_steps = len(series)
print("Total timesteps available:", total_steps)

max_start_idx = total_steps - (PAST_STEPS + FUTURE_STEPS) + 1
if max_start_idx <= 0:
    raise ValueError("Not enough data for the requested window sizes. Reduce PAST_STEPS or FUTURE_STEPS.")

n_windows = max_start_idx
test_windows = int(np.ceil(TEST_RATIO * n_windows))
val_windows = int(np.ceil(VAL_RATIO * n_windows))
train_windows = n_windows - val_windows - test_windows
print(f"Windows total: {n_windows}, train: {train_windows}, val: {val_windows}, test: {test_windows}")

if train_windows <= 0:
    raise ValueError("Train windows <= 0. Reduce test/val ratios or use more data.")

start_indices = np.arange(0, n_windows, dtype=np.int64)

# Optionally limit samples for debugging/speed
if MAX_SAMPLES is not None:
    start_indices = start_indices[:MAX_SAMPLES]
    n_windows = len(start_indices)
    test_windows = int(np.ceil(TEST_RATIO * n_windows))
    val_windows = int(np.ceil(VAL_RATIO * n_windows))
    train_windows = n_windows - val_windows - test_windows
    print(f"Limited windows: {n_windows}, train: {train_windows}, val: {val_windows}, test: {test_windows}")

train_starts = start_indices[:train_windows]
val_starts = start_indices[train_windows:train_windows + val_windows]
test_starts = start_indices[train_windows + val_windows:train_windows + val_windows + test_windows]
print("First few train starts:", train_starts[:5])

# ---------------------------
# 4) Fit scaler on training portion only (avoid leakage)
# ---------------------------
scaler = MinMaxScaler(feature_range=(0, 1))
# Fit scaler on training inputs region to avoid leakage; use up to last input index in training
last_input_idx = train_starts[-1] + PAST_STEPS - 1
fit_slice = series.iloc[: last_input_idx + 1]
scaler.fit(fit_slice.values)
print("Scaler fitted on training portion: ", fit_slice.shape)

# ---------------------------
# 5) Create X, Y arrays from start indices
# ---------------------------
def build_xy_from_starts(series_values, starts, past_steps, future_steps, scaler):
    """
    series_values: numpy array shape (T, 1)
    starts: list/array of start indices (0-based) for inputs: input window is series_values[start : start+past_steps]
    returns: X shape (N, past_steps, n_features), Y shape (N, future_steps)
    """
    n = len(starts)
    X = np.zeros((n, past_steps, series_values.shape[1]), dtype=np.float32)
    Y = np.zeros((n, future_steps), dtype=np.float32)
    for i, s in enumerate(starts):
        in_start = s
        in_end = s + past_steps  # exclusive
        out_start = in_end
        out_end = out_start + future_steps
        window_in = series_values[in_start:in_end]
        window_out = series_values[out_start:out_end].reshape(-1)
        # scale input & output using fitted scaler
        window_in_scaled = scaler.transform(window_in)
        window_out_scaled = scaler.transform(window_out.reshape(-1,1)).reshape(-1)
        X[i] = window_in_scaled
        Y[i] = window_out_scaled
    return X, Y

series_values = series.values  # shape (T, 1)

print("Building train dataset...")
X_train, Y_train = build_xy_from_starts(series_values, train_starts, PAST_STEPS, FUTURE_STEPS, scaler)
print("Building val dataset...")
X_val, Y_val = build_xy_from_starts(series_values, val_starts, PAST_STEPS, FUTURE_STEPS, scaler)
print("Building test dataset...")
X_test, Y_test = build_xy_from_starts(series_values, test_starts, PAST_STEPS, FUTURE_STEPS, scaler)

print("Shapes -> X_train:", X_train.shape, "Y_train:", Y_train.shape)

n_features = X_train.shape[2]

# ---------------------------
# 6) Build the simple MLP model (NAR-like)
# ---------------------------
tf.keras.backend.clear_session()
mlp_model = Sequential([
    Input(shape=(PAST_STEPS, n_features)),
    Flatten(),
    Dense(256, activation='relu'),
    Dense(256, activation='relu'),
    Dense(FUTURE_STEPS, activation='linear')  # predict all FUTURE_STEPS in one shot
])

mlp_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3), loss='mae', metrics=['mae'])
mlp_model.summary()

# ---------------------------
# 7) Training
# ---------------------------
callbacks = [
    EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True),
    ModelCheckpoint(MODEL_SAVE_PATH, monitor='val_loss', save_best_only=True, verbose=1)
]

history = mlp_model.fit(
    X_train, Y_train,
    validation_data=(X_val, Y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks,
    shuffle=False
)

# ---------------------------
# 8) Evaluation - MAE per horizon (minutes ahead)
# ---------------------------
print("Predicting on test set...")
Y_pred_test = mlp_model.predict(X_test)  # shape (N, FUTURE_STEPS)

# Inverse transform outputs back to original scale
Y_test_inv = scaler.inverse_transform(Y_test.reshape(-1, 1)).reshape(-1, FUTURE_STEPS)
Y_pred_inv = scaler.inverse_transform(Y_pred_test.reshape(-1, 1)).reshape(-1, FUTURE_STEPS)

# compute MAE per horizon
mae_per_horizon = []
for h in range(FUTURE_STEPS):
    mae_h = mean_absolute_error(Y_test_inv[:, h], Y_pred_inv[:, h])
    mae_per_horizon.append(mae_h)

overall_mae = mean_absolute_error(Y_test_inv.reshape(-1), Y_pred_inv.reshape(-1))
print(f"Overall test MAE (all horizons): {overall_mae:.4f} kW")

# ---------------------------
# 9) Plots
# ---------------------------
plt.figure(figsize=(10,5))
plt.plot(np.arange(1, FUTURE_STEPS + 1), mae_per_horizon, marker='o', markersize=3)
plt.xlabel('Minutes ahead')
plt.ylabel('MAE (kW)')
plt.title('MAE vs Forecast Horizon (Simple MLP)')
plt.grid(True)
plt.show()

# Plot sample prediction vs actual for a few windows
n_samples_plot = 3
fig, axes = plt.subplots(n_samples_plot, 1, figsize=(10, 3*n_samples_plot))
for i in range(n_samples_plot):
    ax = axes[i]
    idx = i if i < X_test.shape[0] else 0  # ensure within bounds
    input_window = scaler.inverse_transform(X_test[idx].reshape(-1, 1)).reshape(-1)
    actual_future = Y_test_inv[idx]
    pred_future = Y_pred_inv[idx]
    t_input = np.arange(-PAST_STEPS, 0)
    t_future = np.arange(0, FUTURE_STEPS)
    ax.plot(t_input, input_window, label='Input (past 30 min)', marker='o')
    ax.plot(t_future, actual_future, label='Actual future (next 120 min)')
    ax.plot(t_future, pred_future, label='Predicted future')
    ax.axvline(-0.5, color='k', linestyle='--')
    ax.set_xlabel('Minutes (relative)')
    ax.set_ylabel('Global_active_power (kW)')
    ax.legend()
    ax.grid(True)
plt.tight_layout()
plt.show()

# ---------------------------
# 10) Save results
# ---------------------------
np.savez('simple_mlp_results.npz',
         mae_per_horizon=np.array(mae_per_horizon),
         Y_test=Y_test_inv,
         Y_pred=Y_pred_inv)

print("Done. Model saved to:", MODEL_SAVE_PATH)
print("Results saved to simple_mlp_results.npz")
```

```python
# ---------------------------
# LSTM model for multi-step forecasting
# Re-uses arrays: X_train, Y_train, X_val, Y_val, X_test, Y_test, scaler
# and constants: PAST_STEPS, FUTURE_STEPS, BATCH_SIZE, EPOCHS, RANDOM_SEED
# ---------------------------

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

LSTM_MODEL_SAVE_PATH = 'best_lstm.h5'

# reproducibility (already set earlier, but safe to set again)
import tensorflow as tf
tf.random.set_seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)

# clear session to avoid clutter
tf.keras.backend.clear_session()

# Build LSTM model
# Paper config: one hidden layer with 50 memory cells (units), softsign activation, linear output.
lstm_model = Sequential([
    Input(shape=(PAST_STEPS, n_features)),         # n_features should be 1 for your dataset
    LSTM(50, activation='softsign'),              # 50 memory cells as in paper
    # optional dropout to reduce overfitting (tune or remove)
    Dropout(0.1),
    Dense(FUTURE_STEPS, activation='linear')      # predict all FUTURE_STEPS in one shot
])

# Compile with Adagrad optimizer (paper used ADAGRAD). Use MAE loss to match MLP experiment.
lstm_model.compile(optimizer=tf.keras.optimizers.Adagrad(learning_rate=1e-2),
                   loss='mae',
                   metrics=['mae'])

lstm_model.summary()

# Callbacks: early stopping and checkpointing (save best by val_loss)
callbacks = [
    EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True, verbose=1),
    ModelCheckpoint(LSTM_MODEL_SAVE_PATH, monitor='val_loss', save_best_only=True, verbose=1)
]

# Fit model
history_lstm = lstm_model.fit(
    X_train, Y_train,
    validation_data=(X_val, Y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks,
    shuffle=False
)

# ---------------------------
# Evaluation - MAE per horizon
# ---------------------------
print("Predicting on test set (LSTM)...")
Y_pred_test_lstm = lstm_model.predict(X_test)  # shape (N, FUTURE_STEPS)

# Inverse transform outputs back to original scale
Y_test_inv_lstm = scaler.inverse_transform(Y_test.reshape(-1, 1)).reshape(-1, FUTURE_STEPS)
Y_pred_inv_lstm = scaler.inverse_transform(Y_pred_test_lstm.reshape(-1, 1)).reshape(-1, FUTURE_STEPS)

# compute MAE per horizon
from sklearn.metrics import mean_absolute_error
mae_per_horizon_lstm = []
for h in range(FUTURE_STEPS):
    mae_h = mean_absolute_error(Y_test_inv_lstm[:, h], Y_pred_inv_lstm[:, h])
    mae_per_horizon_lstm.append(mae_h)

overall_mae_lstm = mean_absolute_error(Y_test_inv_lstm.reshape(-1), Y_pred_inv_lstm.reshape(-1))
print(f"LSTM overall test MAE (all horizons): {overall_mae_lstm:.4f} kW")

# Plot MAE vs horizon (same plotting style as MLP)
plt.figure(figsize=(10,5))
plt.plot(np.arange(1, FUTURE_STEPS + 1), mae_per_horizon_lstm, marker='o', markersize=3)
plt.xlabel('Minutes ahead')
plt.ylabel('MAE (kW)')
plt.title('MAE vs Forecast Horizon (LSTM)')
plt.grid(True)
plt.show()

# Plot sample predictions (like MLP)
n_samples_plot = 3
fig, axes = plt.subplots(n_samples_plot, 1, figsize=(10, 3*n_samples_plot))
for i in range(n_samples_plot):
    ax = axes[i]
    idx = i if i < X_test.shape[0] else 0
    input_window = scaler.inverse_transform(X_test[idx].reshape(-1, 1)).reshape(-1)
    actual_future = Y_test_inv_lstm[idx]
    pred_future = Y_pred_inv_lstm[idx]
    t_input = np.arange(-PAST_STEPS, 0)
    t_future = np.arange(0, FUTURE_STEPS)
    ax.plot(t_input, input_window, label='Input (past 30 min)', marker='o')
    ax.plot(t_future, actual_future, label='Actual future (next 120 min)')
    ax.plot(t_future, pred_future, label='Predicted future (LSTM)')
    ax.axvline(-0.5, color='k', linestyle='--')
    ax.set_xlabel('Minutes (relative)')
    ax.set_ylabel('Global_active_power (kW)')
    ax.legend()
    ax.grid(True)
plt.tight_layout()
plt.show()

# Save results
np.savez('lstm_results.npz',
         mae_per_horizon=np.array(mae_per_horizon_lstm),
         Y_test=Y_test_inv_lstm,
         Y_pred=Y_pred_inv_lstm)

print("Done. LSTM model saved to:", LSTM_MODEL_SAVE_PATH)
print("Results saved to lstm_results.npz")

# ----------------------------------------------
# EXTRA: Three plots for report (MAE, MSE, VAR)
# ----------------------------------------------

from sklearn.metrics import mean_absolute_error, mean_squared_error

# ============= 1) MAE per forecast horizon =============
mae_per_horizon = []
for h in range(FUTURE_STEPS):
    mae = mean_absolute_error(Y_test_inv_lstm[:, h], Y_pred_inv_lstm[:, h])
    mae_per_horizon.append(mae)

plt.figure(figsize=(10,5))
plt.plot(range(1, FUTURE_STEPS+1), mae_per_horizon, marker='o', markersize=3)
plt.xlabel("Forecast Span [minutes]")
plt.ylabel("MAE (kWh)")
plt.title("Mean Absolute Error (MAE) vs Forecast Span (LSTM)")
plt.grid(True)
plt.show()


# ============= 2) MSE per forecast horizon =============
mse_per_horizon = []
for h in range(FUTURE_STEPS):
    mse = mean_squared_error(Y_test_inv_lstm[:, h], Y_pred_inv_lstm[:, h])
    mse_per_horizon.append(mse)

plt.figure(figsize=(10,5))
plt.plot(range(1, FUTURE_STEPS+1), mse_per_horizon, marker='o', markersize=3)
plt.xlabel("Forecast Span [minutes]")
plt.ylabel("MSE (kWh)")
plt.title("Mean Squared Error (MSE) vs Forecast Span (LSTM)")
plt.grid(True)
plt.show()


# ============= 3) Error Variance per horizon =============
error_variance = []
for h in range(FUTURE_STEPS):
    errors = Y_test_inv_lstm[:, h] - Y_pred_inv_lstm[:, h]
    error_variance.append(np.var(errors))

plt.figure(figsize=(10,5))
plt.plot(range(1, FUTURE_STEPS+1), error_variance, marker='o', markersize=3)
plt.xlabel("Forecast Span [minutes]")
plt.ylabel("Error Variance")
plt.title("Error Variance vs Forecast Span (LSTM)")
plt.grid(True)
plt.show()

print("Generated MAE, MSE, and Error Variance plots for report.")
```
